#### sliding window function

```python
def multivariate_data(dataset, target, start_index, end_index, history_size,
                      target_size, step, single_step=False):
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i-history_size, i, step)
        data.append(dataset[indices])

        if single_step:
            labels.append(target[i+target_size])
        else:
            labels.append(target[i:i+target_size])

    return np.array(data), np.array(labels)
```

### compile and fit function

```python
# Function of training models
MAX_EPOCHS = 20
EVALUATION_INTERVAL = 200
batch_size = 32
buffer_size = 150
def compile_and_fit(model, train, val, patience=2):
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                    patience=patience,
                                                    mode='min') 

    model.compile(loss=tf.keras.losses.MeanSquaredError(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=[tf.keras.metrics.MeanAbsoluteError()])

    history = model.fit(train, batch_size=batch_size, epochs=MAX_EPOCHS,
                      validation_data=val,
                        steps_per_epoch=EVALUATION_INTERVAL,
                      callbacks=[early_stopping],
                    validation_steps=10
                       )
    model.reset_states()
    return history
```

```python
# make a one-step forecast
def forecast(model, batch_size, X, past_history):
    X = X.reshape(1, past_history, num_features)
    yhat = model.predict(X)
    return yhat
```

```python
def inverse_difference(history, yhat, position=1):
    return yhat + history[position]
```

```python
def evaluate_rmse(history, predictions):
    rmse_list=[]
    for i in range(num_features):
        rmse = sqrt(mean_squared_error(history[:,i], predictions[:,i]))
        rmse_list.append(rmse)
    mean_rmse = sum(rmse_list)/len(rmse_list)
    return mean_rmse
```

### Baseline

```python
class Baseline(tf.keras.Model):
  def __init__(self, label_index=None):
    super().__init__()
    self.label_index = label_index

  def call(self, inputs):
    if self.label_index is None:
      return inputs
    result = inputs[:, :, self.label_index]
    return result[:, :, tf.newaxis]

baseline = Baseline()
```

```python
evaluation_val_list = {}
for fold, (train_index, test_index) in enumerate(splits.split(X_single)):
    print("Fold: {}".format(fold))
    x_train_single, x_val_single = X_single[train_index], X_single[test_index]
    y_train_single, y_val_single = y_single[train_index], y_single[test_index]
    print('Observations: %d' % (len(x_train_single) + len(x_val_single)))
    print('Training Observations: %d' % (len(x_train_single)))
    print('Testing Observations: %d' % (len(x_val_single)))
    train_data_single = tf.data.Dataset.from_tensor_slices((x_train_single, y_train_single))
    train_data_single = train_data_single.cache().shuffle(buffer_size).batch(batch_size).repeat()

    val_data_single = tf.data.Dataset.from_tensor_slices((x_val_single, y_val_single))
    val_data_single = val_data_single.batch(batch_size).repeat()
    
    baseline.compile(loss=tf.keras.losses.MeanSquaredError(),
                 metrics=[tf.keras.metrics.MeanAbsoluteError()])
    evaluation_val_list ["Fold {}".format(fold)] = baseline.evaluate(val_data_single, steps=50)



IPython.display.clear_output()
mae = []
# calculate the mean mae
for v in evaluation_val_list.values():
    mae.append(v[1])
Mean_mae = mean(mae)
val_performance['Baseline'] = Mean_mae
test_performance['Baseline'] = baseline.evaluate(test_data_single, verbose=0, steps=50)
```

### LSTM-CNN

#### CNN-LSTM model used the sliding window generated for CNN model

```python
model_cnn_lstm = tf.keras.Sequential([
    # Shape [batch, time, features] => [batch, lstm_units].
    # Adding more `lstm_units` just overfits more quickly.
    tf.keras.layers.Conv1D(filters=512, kernel_size=3, activation='relu', input_shape=(3, num_features)),
    #tf.keras.layers.Conv1D(filters=256, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=1),
    tf.keras.layers.Flatten(),
    tf.keras.layers.RepeatVector(future_target),
    tf.compat.v1.keras.layers.LSTM(200,return_sequences=True),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(100, activation='relu')),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(num_features))
])
```

```python
def walk_forward_validation (model, X_train, X_test, past_history):
    prediction = list()
    expected_history = list ()
    
    for i in range(len(X_test)):
        # make one-step forecast
        X = X_test[i, :]
        yhat = forecast(model, 1, X, past_history)
        if yhat.ndim > 2:
            yhat = yhat[0,:,:]

        # invert scaling
        yhat = scaler.inverse_transform(yhat)
        # invert differencing
        # the inverse differencing is add the yhat and the raw values that is one step ahead than the expected value
        # here we are passing the position of one step ahead raw values to the function
        yhat = inverse_difference(raw_values.values, yhat, len(X_train)+past_history+i)
        # store forecast
        prediction.append(yhat)
        # because of differencing, all the value have shift one step
        expected = raw_values.values[len(X_train)+1+past_history+i]
        expected_history.append(expected)
        #print('Week=%d, Predicted=%f, Expected=%f' % (i+1, yhat[:,8], expected[8]))
    # turn the one dimension list into two dimension
    prediction = np.asarray(prediction)
    prediction = prediction.reshape(len(X_test),num_features)
    return prediction
```
