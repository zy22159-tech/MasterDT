```python
import pandas as pd
import os
import glob
from pathlib import Path
from datetime import datetime
```

```python
p = Path("Data/Commodity")
filelist=list(p.glob("**/*.csv"))
```

```python
dfs_d=[]
for f in filelist:
    df=pd.read_csv(f, sep=',')
    try:
        df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
        df.style.format({"Date": lambda t: t.strftime("%Y-%m-%d")})
    except ValueError:
        pass
    mask = (df['Date'] >'1990-09-01') & (df['Date'] <= '2022-05-01')
    df = df[['Date', 'Price']]
    df = df.loc[mask]
    df.rename(columns = {'Price':f.name}, inplace = True)
    df['Date']=df['Date'].astype("string")
    dfs_d.append(df)
```

```python
dfs_d
```

```python
from functools import reduce
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), dfs_d)
```

```python
df_merged
```

```python
df_merged['Date'] = pd.to_datetime(df_merged['Date'],format='%Y-%m-%d')
df_merged = df_merged.sort_values(by='Date')
```

```python
df_merged.to_csv('Data\combined.csv')
```
