```python
import pandas as pd
import os
import glob
from pathlib import Path
from datetime import datetime
```

```python
p = Path("\Documents\Github\MasterDT\Data\Commodity")
filelist=list(p.glob("**/*.csv"))
```

```python
dfs_d=[]
for f in filelist:
    df=pd.read_csv(f, sep=',')
    df=pd.read_csv(filelist[0], sep=',')
    df['Date']= pd.to_datetime(df['Date'])
    mask = (df['Date'] > '1990-09-01') & (df['Date'] <= '2021-12-31')
    df = df.loc[mask]
    df = df[['Date', 'Price']]
    df.rename(columns = {'Price':f.name}, inplace = True)
    dfs_d.append(df)
```

```python

dfs_d
```

```python
from functools import reduce
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), dfs_d)
```

```python
df_merged.to_csv('Data\combined.csv')
```
