```python
p = Path("\Documents\Master DT\Data")
filelist=list(p.glob("**/*.txt"))
```

```python
dfs=[]
for file in filelist:
    f = open(file,'r')
    row=[]
    data = f.readlines()
    for line in data:
        if not line.strip("\n").startswith('#'):
            line = line.rstrip('\n')
            row.append(line)
    row=row[1:]
    z = [i.split(',') for i in row]
    df = pd.DataFrame(z, columns=['Date','station_num','station_name','SWE'])
    dfs.append(df)
```

```python
SWE_merged = pd.concat(dfs, ignore_index=True)
```

```python
SWE_merged['SWE']=pd.to_numeric(SWE_merged['SWE'])

SWE_merged.drop(columns=['station_num', 'station_name'])
```

```python
SWE_merged=SWE_merged.groupby(SWE_merged['Date']).sum()
```

```python
SWE_merged.to_csv('SWE_sum.csv') 
```

```python
p = Path("\Documents\Github\MasterDT\Data\canadian")
filelist=list(p.glob("**/*.csv"))
```

```python
df_from_each_file = (pd.read_csv(f, sep=',') for f in filelist)
ca_merged   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
ca_merged
```

```python
ca_new=pd.DataFrame()
ca_new=ca_merged.groupby(ca_merged['LOCAL_DATE']).agg({'MEAN_TEMPERATURE': lambda x: x.mean(skipna=True)})
ca_new['min_humidity']=ca_merged.groupby(ca_merged['LOCAL_DATE']).agg({'MIN_REL_HUMIDITY': lambda x: x.mean(skipna=True)})
ca_new['max_humidity']=ca_merged.groupby(ca_merged['LOCAL_DATE']).agg({'MAX_REL_HUMIDITY': lambda x: x.mean(skipna=True)})
ca_new['precipitation']=ca_merged.groupby(ca_merged['LOCAL_DATE']).agg({'TOTAL_PRECIPITATION': lambda x: x.sum()})
```

```python
p = Path("\Documents\Github\MasterDT\Data\Canadian_climate_history.csv")
ca_1=pd.read_csv(p, sep=',')
```
