```python
p = Path("\Documents\Master DT\Data")
filelist=list(p.glob("**/*.txt"))
```

```python
dfs=[]
for file in filelist:
    f = open(file,'r')
    row=[]
    data = f.readlines()
    for line in data:
        if not line.strip("\n").startswith('#'):
            line = line.rstrip('\n')
            row.append(line)
    row=row[1:]
    z = [i.split(',') for i in row]
    df = pd.DataFrame(z, columns=['Date','station_num','station_name','SWE'])
    dfs.append(df)
```

```python
SWE_merged = pd.concat(dfs, ignore_index=True)
```

```python
SWE_merged['SWE']=pd.to_numeric(SWE_merged['SWE'])

SWE_merged.drop(columns=['station_num', 'station_name'])
```

```python
SWE_merged=SWE_merged.groupby(SWE_merged['Date']).sum()
```

```python
SWE_merged.to_csv('SWE_sum.csv') 
```
