```python
import pandas as pd
import os
import glob
from pathlib import Path
```

```python
p = Path("/Users/wangz/Documents/snow")
 
filelist=list(p.glob("**/*.csv"))
```

```python
df_from_each_file = (pd.read_csv(f, sep=',') for f in filelist)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
us_new=df_merged.groupby(df_merged['Date']).agg({'tmax': lambda x: x.mean(skipna=True)})
us_new['tmin']=df_merged.groupby(df_merged['Date']).agg({'tmin': lambda x: x.mean(skipna=True)})
us_new['prcp']=df_merged.groupby(df_merged['Date']).agg({'prcp': lambda x: x.sum()})
```

```python
us_new.reset_index(inplace=True)
```

```python
us_new['Date'] =  pd.to_datetime(us_new['Date'])
mask = (us_new['Date'] >'1990-08-31') & (us_new['Date'] <= '2022-05-01')
us_new = us_new.loc[mask]
```

```python
us_new['mean_temp'] = us_new.iloc[:, [1,2]].mean(skipna=True,axis=1)
```

```python
us_new.to_csv('Data/us_Tem_pre.csv') 
```

```python
p = Path("\Documents\Master DT\Data")
filelist=list(p.glob("**/*.txt"))
```

```python
dfs=[]
for file in filelist:
    f = open(file,'r')
    row=[]
    data = f.readlines()
    for line in data:
        if not line.strip("\n").startswith('#'):
            line = line.rstrip('\n')
            row.append(line)
    row=row[1:]
    z = [i.split(',') for i in row]
    df = pd.DataFrame(z, columns=['Date','station_num','station_name','SWE'])
    dfs.append(df)
```

```python
SWE_merged = pd.concat(dfs, ignore_index=True)
```

```python
SWE_merged['SWE']=pd.to_numeric(SWE_merged['SWE'])

SWE_merged.drop(columns=['station_num', 'station_name'])
```

```python
SWE_merged=SWE_merged.groupby(SWE_merged['Date']).sum()
```

```python
SWE_merged.to_csv('SWE_sum.csv') 
```

```python
df_from_each_file = (pd.read_csv(f, sep=',') for f in filelist)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
df_Snow_file = []
for f in filelist:
    df = pd.read_csv(f, sep=',') 
    if set(['SNOW','SNWD']).issubset(df.columns):
        df_Snow_file.append(df)
```

```python
df_Snow_file = pd.concat(df_Snow_file, ignore_index=True)
```

```python
snow=df_Snow_file.groupby(df_Snow_file['DATE']).agg({'SNOW': lambda x:  x.sum()})
snow['SNWD']=df_Snow_file.groupby(df_Snow_file['DATE']).agg({'SNWD': lambda x: x.mean(skipna=True)})
```

```python
snow=snow.reset_index()
```

```python
snow.to_csv('snow.csv')
```
